```python
import seaborn as sns
%matplotlib inline

sns.get_dataset_names()
```

```python
tips = sns.load_dataset('tips')
flights = sns.load_dataset('flights')
tips.head()
```

```python
flights.head(20)
```

```python
# Data should be in a matrix form labels on both axis should be the same; 
# that is, a pivot table

# Correlation data
tips_corr = tips.corr()
```

```python
sns.set_context('talk', font_scale =1)
sns.heatmap(tips_corr)
```

```python
tips_corr
```

```python
sns.heatmap(tips_corr, annot=True)
```

```python
sns.heatmap(tips_corr, annot=True, cmap='coolwarm')
```

```python
# Now let's turn 'flights' dataset into a pivot table

fp = flights.pivot_table(index='month',columns='year',values='passengers')
fp
```

```python
sns.heatmap(fp)
```

```python
sns.heatmap(fp,cmap='viridis',linecolor='black', linewidth=0.1)
```

```python
# Cluster map

sns.clustermap(fp,cmap='coolwarm')
```

```python
# Normalize the flight scale (make it from 0 to 1)

sns.clustermap(fp,cmap='coolwarm',standard_scale=1)
```

```python
sns.set_context('talk', font_scale =0.8)

sns.lmplot(x='total_bill',y='tip',data=tips,hue='sex',palette='viridis')
```

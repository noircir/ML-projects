```python
import seaborn as sns
%matplotlib inline
tips = sns.load_dataset('tips')
tips.head()
```

```python
# x is categorical attribute, y is a numeric attribute, so it doesn't need to be passed
# By default, estimator = the average (mean) of y for each of x.

sns.set_style('whitegrid')
sns.barplot(x='sex', y='total_bill', data=tips)
```

```python
# What if we would like to know not a mean, but a standard deviation of total bill by gender?
import numpy as np

sns.barplot(x='sex', y='total_bill', data=tips, estimator=np.std)
```

```python
# Countplot (the estimator is counting the number of occurences)
# More Dinner in dataset

sns.countplot(x='time', data=tips)
```

```python
# More Males in dataset
sns.countplot(x='sex', data=tips)
```

```python
# More non-smokers in dataset

sns.countplot(x='smoker', data=tips)
```

```python
# More Saturdays in dataset

sns.countplot(x='day', data=tips)
```

```python
# 'size' is a numerical value, but categorical (not continuous), so can do a countplot
# more 2-people party

sns.countplot(x='size', data=tips)
```

## Boxplot

```python
# The boxplots show the quartiles of a dataset. The whiskers show 
# the rest of the distribution. The points outside of the whiskers are outliers.

sns.boxplot(x='day', y='total_bill', data=tips)
```

```python
# Split the boxplot even further by adding another category "smoker"
sns.boxplot(x='day', y='total_bill', data=tips,hue='smoker')
```

## Violin plots

```python
# Violin plots take into account all points, even outliers.
# Their curve is KDE on the sides.

sns.violinplot(x='day', y='total_bill', data=tips)
```

```python
# Add another category

sns.violinplot(x='day', y='total_bill', data=tips,hue='sex')
```

```python
# Condense the violins: one side is one category, the other side is the other category.

sns.violinplot(x='day', y='total_bill', data=tips,hue='sex', split=True)
```

## Strip plot

```python
sns.stripplot(x='day', y='total_bill', data=tips)
```

```python
sns.stripplot(x='day', y='total_bill', data=tips, jitter=True)
```

```python
sns.stripplot(x='day', y='total_bill', data=tips, hue='sex')
```

```python
sns.stripplot(x='day', y='total_bill', data=tips, hue='sex', dodge=True)
```

```python
# Swarm plot - a combination of violin plot and strip plot.

sns.swarmplot(x='day', y='total_bill', data=tips)
```

```python
sns.swarmplot(x='day', y='total_bill', data=tips, hue='sex')
```

```python
sns.swarmplot(x='day', y='total_bill', data=tips, hue='sex', dodge=True)
```

```python
# Combining swarm plot with violin plot

sns.violinplot(x='day', y='total_bill', data=tips)
sns.swarmplot(x='day', y='total_bill', data=tips,color='black')
```

## Factor plot

```python
sns.catplot(x='day', y='total_bill', data=tips,kind='bar')
```

```python
sns.catplot(x='day', y='total_bill', data=tips,kind='violin')
```
